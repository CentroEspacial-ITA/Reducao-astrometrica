```python
import numpy as np
from astropy.io import fits
from skimage import exposure, filters, measure, morphology
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

def detect_satellite_trail(image_path, threshold):
    # Carregando a imagem FITS
    fits_file = fits.open(image_path)
    image_obs = fits_file[0].data[0]
    fits_file.close()

    # Pré-processamento
    image_normalized = exposure.equalize_hist(image_obs)
    image_blurred = filters.gaussian(image_normalized, sigma=1)

    # Detecção de borda
    edges = filters.sobel(image_blurred)

    # Invertendo a imagem para destacar as regiões escuras
    inverted_edges = 1 - edges
    
    # Limiarização para criar uma imagem binária
    binary_image = inverted_edges > 0.95

    # Remoção de objetos pequenos
    binary_image = morphology.remove_small_objects(binary_image, min_size=threshold)

    # Identificando contornos
    labeled_regions = measure.label(binary_image)
    regions = measure.regionprops(labeled_regions)

    # Filtragem de regiões
    filtered_regions = [region for region in regions if region.eccentricity > 0.95]

    # Encontrar coordenadas de início, meio e fim da trajetória
    results = []
    for region in filtered_regions:
        minr, minc, maxr, maxc = region.bbox
        start_point = (minc, minr)
        end_point = (maxc, maxr)
        coords = np.column_stack(np.where(binary_image[minr:maxr, minc:maxc]))
        middle_point = coords[int(coords.shape[0] / 2)]
        middle_point = (middle_point[1] + minc, middle_point[0] + minr)
        results.append((start_point, middle_point, end_point))

    reta = [[coord[1]+ minc, coord[0]+minr] for coord in coords]

    # Suavização da curva
    smoothed_curve = gaussian_filter1d(np.array(reta), sigma=5, axis=0)

    # Visualização
    fig, axes = plt.subplots(1, 4, figsize=(20, 5))

    axes[0].imshow(image_obs, cmap='gray')
    axes[0].set_title('Original Image')

    axes[1].imshow(edges, cmap='gray')
    axes[1].set_title('Edges')

    axes[2].imshow(binary_image, cmap='gray')
    axes[2].set_title('Binary Image')

    axes[3].imshow(image_obs, cmap='gray')
    for start, middle, end in results:
        axes[3].plot([start[0], middle[0], end[0]], [start[1], middle[1], end[1]], '-r', linewidth=2)
        axes[3].scatter([start[0], middle[0], end[0]], [start[1], middle[1], end[1]], s=25, c='g')
    axes[3].plot(smoothed_curve[:, 0], smoothed_curve[:, 1], '-r', linewidth=2)
    axes[3].set_xlim((0, image_obs.shape[1]))
    axes[3].set_ylim((image_obs.shape[0], 0))
    axes[3].set_title('Smoothed Curve')

    plt.tight_layout()
    plt.show()

    # Resultados
    for i, (start, middle, end) in enumerate(results):
        print(f"Region {i+1} - Start Point: {start}, Middle Point: {middle}, End Point: {end}")

    return smoothed_curve

# Substitua 'path_to_your_fits_file.fits' pelo caminho para o seu arquivo FITS
threshold = 22000
smoothed_curve = detect_satellite_trail('./Images/NOSS 3-4A230629190753.fit', threshold)
```
