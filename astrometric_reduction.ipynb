```python
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
fits_file = fits.open('./Images/J75_01_20120824040153.fits')
print(fits_file)
```

```python
image_data = fits_file[0].data  #brilho de cada pixel da imagem
print(image_data)
```

```python
plt.figure()
plt.imshow(image_data, origin = 'lower', norm = LogNorm(), cmap = 'Greys')
plt.colorbar()
plt.show()
```

```python
section1 = image_data[1300:1600, 1800:2000]

plt.figure()
plt.imshow(section1, origin = 'lower', norm = LogNorm(), cmap = 'Greys')
plt.colorbar()
plt.show()
```
